````python
from collections import Counter 
from nltk.tokenize import word_tokenize
import pandas as pd
import nltk
````

````python
data1 = word_tokenize(data)
````

````python
data1
````

````python
len(data1)
````

````python
Counter(data1).most_common(10)
````

````python
# Remove punctuations 
data2 = [ x for x in data1 if x.isalpha()]
````

````python
data2
````

````python
# making lowercase 
data3 = [ x.lower() for x in data2 if x.isalpha()]
````

````python
data3
````

````python
Counter(data3).most_common(10)
````

````python
# remove stopwards 
from nltk.corpus import stopwords
````

````python
stoward_list =stopwords.words('english')
stoward_list
````

````python
data4 = [x for x in data3 if x not in stoward_list]
````

````python
Counter(data4).most_common(10)
````

#### Stemming and lemmatization both generate the root form of the words. The difference is that stem might not be actual word whereas lemma is actual language word

````python
# stemming and lemmatizer 

from nltk.stem import LancasterStemmer,WordNetLemmatizer
````

````python
# stemming
lst = LancasterStemmer()
# lst.stem('theory')



words =['friend','friendship','friends','stabil','destabilize','football']
for i in words:
    print(lst.stem(i))
````

````python
lst.stem('mathematician')
````

````python
wnl = WordNetLemmatizer()
for i in words :
    print(wnl.lemmatize(i))
````

````python
# portertemmer
````

````python
# combining roots 
data = [wnl.lemmatize(i) for i  in data4]
````

````python
data4
````

````python
data
````

````python
Counter(data).most_common(10)
````
