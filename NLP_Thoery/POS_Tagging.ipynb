```python
# check which are most used POS 
nltk.download('brown')
from nltk.corpus import brown


brown_tagged_words = brown.tagged_words(tagset='universal',categories='news')
```

```python
brown_tagged_words
```

```python
freqdist = nltk.FreqDist([tag for (word,tag) in brown_tagged_words])
```

```python
freqdist.most_common(10)
```

```python
# check bigrams 

bigrams = nltk.bigrams(brown_tagged_words)
```

```python
bigrams
```

```python
bigram = [w for w in bigrams]
```

```python
bigram
```

```python
# check noun preceded bigrams 
noun_preceded = [w1 for (w1,w2) in bigram if w2[1]=='NOUN' ]
```

```python
noun_preceded
```

```python
# check bigram frequency preceded with noun
Bigram_frequency = nltk.FreqDist(noun_preceded)
Bigram_frequency.most_common()
```
