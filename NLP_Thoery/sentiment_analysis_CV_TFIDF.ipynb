```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\DELL\Documents\NLP\Train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.text,df.label,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

## Count_Vectorizer

```python
# oov out of vocabulary
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
max_words = 10000
cv = CountVectorizer(lowercase=True,stop_words='english',max_df=0.95,max_features = max_words)
```

```python
count_train= cv.fit_transform(x_train)
count_test = cv.transform(x_test)
```

```python
count_train
```

```python
print(cv.get_stop_words())
```

```python
print(cv.get_feature_names())
```

```python
pd.DataFrame(count_train.A,columns = cv.get_feature_names()).head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
mnb = MultinomialNB()
mnb.fit(count_train.A,y_train)
pred_mnb = mnb.predict(count_test.A)
```

```python
acc = accuracy_score(y_test,pred_mnb)
acc
```

# TF-IDF_Vectorizer 


```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
max_words = 10000
tf = TfidfVectorizer(lowercase=True,stop_words='english',max_df=0.95,max_features = max_words)
```

```python
tf_train= tf.fit_transform(x_train)
tf_test = tf.transform(x_test)
```

```python
pd.DataFrame(tf_train.A,columns = tf.get_feature_names()).head()
```

```python
mnb = MultinomialNB()
mnb.fit(tf_train.A,y_train)
pred_mnb = mnb.predict(tf_test.A)
```

```python
acc = accuracy_score(y_test,pred_mnb)
acc
```
